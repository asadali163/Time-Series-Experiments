```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

##### First let's study the PACF plot

```python
x0 = np.random.randn(1000)
```

```python
plt.plot(x0);
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x0, ax=ax);
```

Let's create AR1 chart

```python
x1 = [0]
for i in range(1000):
  x = (0.5 * x1[-1]) + (0.1 * np.random.randn())
  x1.append(x)
x1 = np.array(x1)
```

```python
plt.plot(x1);
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x1, ax=ax);
```

Let's create AR2 chart

```python
x2 = [0,0]
for i in range(1000):
  x = (0.5 * x2[-1]) + (0.2 * x2[-2]) + (0.1 * np.random.randn())
  x2.append(x)
x2 = np.array(x2)
```

```python
plt.plot(x2)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x2, ax=ax);
```

Now let's create AR5 chart that only depends on lag1, lag2, and lag5. The middle lags like lag 3, and lag4 will still have affect on it

```python
x5 = [0, 0, 0, 0, 0]
for i in range(1000):
  x = (0.5 * x5[-1]) - (0.4 * x5[-2]) + (0.3 * x5[-5])  + (0.1 * np.random.randn())
  x5.append(x)
x5 = np.array(x5)
```

```python
plt.plot(x5);
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x5, ax=ax);
```

### Let's work with ACF plots for MA

```python
### iid noise
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(x0, ax=ax);
```

Let's try MA1 plot with some random noise

```python
errors = 0.1 * np.random.randn(1000)
ma1 = []
for i in range(1000):
  x = (0.5 * errors[i-1]) + errors[i]
  ma1.append(x)
ma1 = np.array(ma1)
```

```python
plt.plot(ma1);
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(ma1, ax=ax);
```

Let's try MA2 Plot

```python
ma2 = []
for i in range(1000):
  x = (0.5 * errors[i-1]) + (0.6 * errors[i-2]) + errors[i]
  ma2.append(x)
ma2 = np.array(ma2)
```

```python
plt.plot(ma2);
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(ma2, ax=ax);
```

Let's try MA3 plot

```python
plt.plot(ma3)
```
